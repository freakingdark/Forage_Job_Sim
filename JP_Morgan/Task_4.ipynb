```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv(r"D:\data_projects\Task 3 and 4_Loan_Data.csv")

X = df[['fico_score']].astype(float)   
y = df['default'].astype(int)
r = 20  
tree = DecisionTreeClassifier(max_leaf_nodes=r, min_samples_leaf=20)
tree.fit(X, y)

thresholds = np.sort(tree.tree_.threshold[tree.tree_.threshold != -2])

print("Optimal Split Thresholds (FICO bins):", thresholds)

bins = np.concatenate(([-np.inf], thresholds, [np.inf]))
df['fico_bin'] = np.digitize(df['fico_score'], bins) - 1

bin_summary = df.groupby('fico_bin').agg(
    count=('default', 'size'),
    defaults=('default', 'sum'),
    avg_fico=('fico_score', 'mean')
)
bin_summary['default_rate'] = bin_summary['defaults'] / bin_summary['count']

print("\nBin Summary:")
print(bin_summary)
```
